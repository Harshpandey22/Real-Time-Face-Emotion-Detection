## Visualization

### Histogram for Percentage Distribution

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['percent'], bins=20, kde=True)
plt.title('Percentage Distribution')
plt.xlabel('Percentage')
plt.ylabel('Frequency')
plt.show()
```

### Bar Plot for Gender Distribution

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='gender', data=df)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

### Time Series Plot for Percentage Over Years

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='year', y='percent', data=df)
plt.title('Percentage Over Years')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.show()
```

### Box Plot for Percentage Distribution by Gender

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='percent', data=df)
plt.title('Percentage Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Percentage')
plt.show()
```

### Scatter Plot for Percentage vs Year

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='year', y='percent', data=df)
plt.title('Percentage vs Year')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.show()
```

### Grouped Bar Plot for Percentage by Gender and Year

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='year', y='percent', hue='gender', data=df)
plt.title('Percentage by Gender and Year')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.legend(title='Gender')
plt.show()
```

### Pie Chart for Gender Distribution

```python
gender_counts = df['gender'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Gender Distribution')
plt.show()
```

### Bar Plot for Top 10 Names

```python
plt.figure(figsize=(10, 6))
top_names = df['name'].value_counts().nlargest(10)
top_names.plot(kind='bar', color=sns.color_palette("husl", 10))
plt.title(f'Top {10} Names')
plt.xlabel('Name')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Bar Plot for Gender vs Percentage

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='gender', y='percent', data=df, errorbar=None, palette='Set2')
plt.title('Gender vs. Percentage')
plt.xlabel('Gender')
plt.ylabel('Average Percentage')
plt.show()
```

### Bar Plot for Top 5 Names vs Percentage Distribution

```python
top_n_names = 5  # Number of top names to display
top_names = df['name'].value_counts().nlargest(top_n_names).index
plt.figure(figsize=(10, 6))
sns.barplot(x='name', y='percent', data=df[df['name'].isin(top_names)], errorbar=None, palette='Paired')
plt.title(f'Top {top_n_names} Names vs Percentage')
plt.xlabel('Name')
plt.ylabel('Average Percentage')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Line Plot for Gender Percentage Trends Over Years

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='year', y='percent', hue='gender', data=df)
plt.title('Gender Percentage Trends Over Years')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Pairplot for Numerical Columns

```python
sns.pairplot(df[numerical_cols])
plt.show()
```

### Coorelation HeatMap

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
